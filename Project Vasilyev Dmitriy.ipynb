Из информации о таблице видно, что в столбцах LifeSquare и Healthcare_1 имеются пропуски.
Для начала разберемся с пропусками в LifeSquare. Так как в квартире не может быть жилой площади равной 0, то нулями заполнять пропуски нельзя. Так же не корректно заполнять пропуски средним значением, так как может получиться ситуация, когда жилая площадь будет больше, чем общая. Теоретически, жилая площадь может равняться общей (например, студия из 1 комнаты, но нам не известен тип жилья). Выведем усредненный коэфициент отношения жилой площади к общей и воспользуемся им для вычисления значения жилой площади от общей в строках с пропущенными значениями.

```python
df_train['LifeSquare'].fillna(df_train['LifeSquare'].median() / 
                              df_train['Square'].median() * df_train['Square'], inplace = True)
```

```python
df_test['LifeSquare'].fillna(df_train['LifeSquare'].median() / 
                              df_train['Square'].median() * df_test['Square'], inplace = True)
```

```python
df_train['LifeSquare'].isna().sum(), df_test['LifeSquare'].isna().sum()
```

Столбец Healthcare более правильно перевести, как здравоохранение и он, вероятно, характеризует наличие больниц рядом с квартирой.
Заполним пропуски в колонке Helthcare_1 средним по медиане.

```python
df_train['Healthcare_1'].fillna(df_train['Healthcare_1'].median(), inplace = True)
```

```python
df_test['Healthcare_1'].fillna(df_train['Healthcare_1'].median(), inplace = True)
```

```python
df_train['Healthcare_1'].isna().sum(), df_test['Healthcare_1'].isna().sum()
```

Мы избавились от пропусков значений в тренировочном и тестовом датафреймах. Теперь необходимо перевести категорийные значения в столбцах Ecology_2, Ecology_3, Shops_2 в последовательные.

```python
replace_col = ['Ecology_2', 'Ecology_3', 'Shops_2']
df_train[replace_col] = df_train[replace_col].replace({'A':0, 'B':1})
df_test[replace_col] = df_test[replace_col].replace({'A':0, 'B':1})
```

### Построение предсказательной модели

Из df_train сформируем тренировочную и тестовую выборку для проверки точности предсказаний.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, GridSearchCV, cross_val_score
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split
```

```python
def evaluate_preds(true_values_train, pred_values_train, true_values_test, pred_values_test):
    print('Train:\t' + 'R2 = ' + str(round(r2_score(true_values_train, pred_values_train), 3)) + 
          '\tMAE = ' + str(int(mean_absolute_error(true_values_train, pred_values_train))) +
          '\n' +
          'Test:\t' + 'R2 = ' + str(round(r2_score(true_values_test, pred_values_test), 3)) +
          '\tMAE = ' + str(int(mean_absolute_error(true_values_test, pred_values_test)))
         )
    
    plt.figure(figsize=(16, 6))
       
    plt.subplot(121)
    sns.scatterplot(x=pred_values_train, y=true_values_train)
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('Train: True vs Predicted values');
    
    plt.subplot(122)
    sns.scatterplot(x=pred_values_test, y=true_values_test)
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('Test: True vs Predicted values');
```

```python
y_train_preds = lr_model.predict(X_train)
y_test_preds = lr_model.predict(X_test)
```

```python
evaluate_preds(y_train, y_train_preds, y_test, y_test_preds)
```
